```python
import os
import json
import re
import warnings
warnings.filterwarnings('ignore')

# ─── CONFIGURE THIS PATH to your audio folder ───
AUDIO_FOLDER = ''   # Folder containing .wav / .mp3 / .m4a files
OUTPUT_FOLDER = './meeting_outputs' # Where results will be saved
WHISPER_MODEL = 'base'              # Options: tiny | base | small | medium | large
                                    # 'base' is best for interview demos (fast + accurate)

os.makedirs(OUTPUT_FOLDER, exist_ok=True)

# List available audio files
audio_extensions = {'.wav', '.mp3', '.m4a', '.flac', '.ogg', '.mp4'}
audio_files = [
    f for f in os.listdir(AUDIO_FOLDER)
    if os.path.splitext(f)[1].lower() in audio_extensions
]

print(f'📂 Audio folder: {os.path.abspath(AUDIO_FOLDER)}')
print(f'📁 Output folder: {os.path.abspath(OUTPUT_FOLDER)}')
print(f'\n🎵 Found {len(audio_files)} audio file(s):')
for f in sorted(audio_files):
    size_kb = os.path.getsize(os.path.join(AUDIO_FOLDER, f)) / 1024
    print(f'   • {f}  ({size_kb:.1f} KB)')
```
